# Import Liberary

```python
import tensorflow as tf
from keras.preprocessing.image import ImageDataGenerator
```

# Data Preparation

```python
train_datagen=ImageDataGenerator(rescale=1./255,
                                shear_range=0.2,
                                zoom_range=0.2,
                                horizontal_flip=True)
training_set=train_datagen.flow_from_directory('../input/brain-tumor-classification-mri/Training',
                                              target_size=(64,64),
                                              batch_size=30,
                                              class_mode='categorical')
```

```python
test_datagen=ImageDataGenerator(rescale=1./255)
testing_set=test_datagen.flow_from_directory('../input/brain-tumor-classification-mri/Training',
                                              target_size=(64,64),
                                              batch_size=30,
                                              class_mode='categorical')
```

# Building CNN

```python
cnn=tf.keras.Sequential()
```

```python
cnn.add(tf.keras.layers.Conv2D(filters=32,kernel_size=3,activation='relu',input_shape=[64,64,3]))
```

```python
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2,strides=2))
```

```python
cnn.add(tf.keras.layers.Conv2D(filters=32,kernel_size=3,activation='relu'))
```

```python
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2,strides=2))
```

```python
cnn.add(tf.keras.layers.Conv2D(filters=32,kernel_size=3,activation='relu'))
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2,strides=2))
```

```python
cnn.add(tf.keras.layers.Flatten())
```

# Full Connection


```python
cnn.add(tf.keras.layers.Dense(units=150,activation='relu'))
```

# Output Layer

```python
cnn.add(tf.keras.layers.Dense(units=4,activation='softmax'))
```

# Compiling CNN

```python
cnn.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

# Training CNN on train data and testing on test data

```python
cnn.fit(x=training_set,validation_data=testing_set,epochs=30)
```
